csv version

```python
modelFile = open('model_result.csv', "r")
model_reader = csv.reader(modelFile)
policyFile = open('policy_result.csv', 'r')
policy_reader = csv.reader(policyFile)

model_result = []
policy_result = []
policy_time = []
for item in model_reader:
    # 忽略第一行
    if model_reader.line_num == 1:
        continue
    model_result.append(item[2])

for item in policy_reader:
    # 忽略第一行
    if policy_reader.line_num == 1:
        continue
    policy_result.append(item[1])
    policy_time.append(item[2])
```

```python
l1=plt.plot(policy_time,policy_result,'#C00000',label='Decision tables')
l2=plt.plot([i for i in range(len(model_result))],model_result,'#2F5597',label='DeepDLP')
plt.title('Pricing process')
plt.xlabel('Step')
plt.ylabel('Price')
plt.legend()
#plt.grid()
plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.plot(policy_time,policy_price,'r--',label='policy')
ax2=ax1.twinx()
ax2.bar(policy_time,policy_sale)
plt.title('The Lasers in Three Conditions')
plt.xlabel('row')
plt.ylabel('column')
plt.legend()
plt.show()
```
